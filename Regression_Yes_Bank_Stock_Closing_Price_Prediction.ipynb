## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

**Statement 1:** The closing price of the nth month is correlated to the opening price of the (n+1)th month.

**Statement 2:** The average volatility has increased after the 2018 fraud case.

**Statement 3:** The average closing price of Yes Bank stock is significantly different before and after the 2018 fraud case.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Statement:** The closing price of the nth month is correlated to the opening price of the (n+1)th month.

**Null Hypothesis (H0):** There is no significant correlation between the closing price of the nth month and the opening price of the (n+1)th month.

**Alternative Hypothesis (H1):** There is a significant correlation between the closing price of the nth month and the opening price of the (n+1)th month.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Create a new DataFrame to store nth month closing prices and (n+1)th month opening prices
df_shifted = pd.DataFrame()
df_shifted['Close_n'] = new_df['Close']
df_shifted['Open_n+1'] = new_df['Open'].shift(-1)

# Drop the last row since it will have NaN in 'Open_n+1'
df_shifted.dropna(inplace=True)

# Calculate the Pearson correlation coefficient
correlation, p_value = stats.pearsonr(df_shifted['Close_n'], df_shifted['Open_n+1'])

print("Correlation Coefficient:", correlation)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant correlation between the closing price of the nth month and the opening price of the (n+1)th month.")
else:
    print("Fail to reject the null hypothesis: There is no significant correlation.")
````

##### Which statistical test have you done to obtain P-Value?

I performed Pearson's correlation coefficient test to obtain the P-value.

##### Why did you choose the specific statistical test?

Pearson's correlation coefficient test is ideal for measuring the linear relationship between two continuous variables, making it suitable for analyzing the correlation between closing and opening stock prices across months.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Statement :** The average volatility has increased after the 2018 fraud case.

**Null Hypothesis (H0):** The average volatility before and after the 2018 fraud case is the same.

**Alternative Hypothesis (H1):** The average volatility after the 2018 fraud case is higher.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Copying Dataset
df_vol = new_df.copy()

# Define the fraud case date
fraud_case_date = pd.to_datetime('2018-01-01')

# Calculate volatility as the difference between high and low prices
df_vol['Volatility'] = df_vol['High'] - df_vol['Low']

# Split the data into before and after the fraud case
volatility_before = df_vol[df_vol.index < fraud_case_date]['Volatility']
volatility_after = df_vol[df_vol.index >= fraud_case_date]['Volatility']

# Perform a two-sample t-test (one-tailed test)
t_stat, p_value = stats.ttest_ind(volatility_before, volatility_after, alternative='less')

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: The average volatility has increased after the 2018 fraud case.")
else:
    print("Fail to reject the null hypothesis: There is no significant increase in volatility.")
````

##### Which statistical test have you done to obtain P-Value?

I performed a two-sample t-test to obtain the P-value.

##### Why did you choose the specific statistical test?

I chose the two-sample t-test because it is suitable for comparing the means of two independent groups, in this case, the volatility before and after the 2018 fraud case, to determine if there is a significant increase in volatility.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Statement:** The average closing price of Yes Bank stock is significantly different before and after the 2018 fraud case.

**Null Hypothesis (H0):** The average closing price before and after the 2018 fraud case is the same.

**Alternative Hypothesis (H1):** The average closing price before and after the 2018 fraud case is different.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Copying Dataset
df_copy = new_df.copy()

# Define the fraud case date
fraud_case_date = pd.to_datetime('2018-01-01')

# Split the data into before and after the fraud case
closing_before = df_copy[df_copy.index < fraud_case_date]['Close']
closing_after = df_copy[df_copy.index >= fraud_case_date]['Close']

# Perform a two-sample t-test (two-tailed test)
t_stat, p_value = stats.ttest_ind(closing_before, closing_after)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: The average closing price is significantly different before and after the 2018 fraud case.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average closing price.")
````

##### Which statistical test have you done to obtain P-Value?

I performed a two-sample t-test to obtain the P-value.

##### Why did you choose the specific statistical test?

I chose the two-sample t-test because it is appropriate for comparing the means of two independent groups, specifically the average closing prices before and after the 2018 fraud case, to determine if there is a significant difference.

In a linear regression model, we assume a linear relationship between the dependent variable (Close) and the independent variables. This implies that changes in the dependent variable are directly proportional to changes in the independent variables.

During training, the linear regression model aims to determine coefficients that minimize the sum of squared differences between predicted and actual values of the dependent variable.

**MSE (Mean Squared Error):**

MSE quantifies the average squared difference between predicted and actual values. Lower MSE indicates more precise predictions, critical for minimizing financial losses and optimizing investment strategies.

**RMSE (Root Mean Squared Error):**

RMSE is the square root of MSE and measures the average magnitude of prediction errors. It provides a clear understanding of the typical deviation of predicted stock prices from actual values, aiding risk assessment and decision-making.

**R2 (R-squared):**

R2 indicates the proportion of variance in stock prices explained by the model. A higher R2 signifies stronger predictive power, offering confidence in the model's ability to explain and forecast stock price movements accurately.

**Adjusted R2 (Adjusted R-squared):**

Adjusted R2 adjusts R2 for the number of predictors in the model, providing a more realistic assessment of its explanatory power. It penalizes excessive complexity, ensuring that improvements in R2 are meaningful and not due to overfitting. Higher Adjusted R2 reflects a robust model with balanced complexity and predictive accuracy.

**MAE (Mean Absolute Error):**

MAE measures the average absolute difference between predicted and actual values. It provides a straightforward assessment of prediction accuracy without considering the direction of errors. Lower MAE suggests better accuracy in predicting stock prices, enhancing the reliability of financial forecasts and investment strategies.

MSE measures the average squared difference between the actual and predicted values. It is crucial for understanding how well the model captures the true stock price movements. A lower MSE indicates that the model's predictions are close to the actual values, which is essential for making accurate financial decisions. For a bank's stock price prediction, where even small deviations can result in significant financial implications, minimizing MSE helps in maintaining the reliability of the predictions.

2. R² (Coefficient of Determination):
R² explains the proportion of the variance in the dependent variable that is predictable from the independent variables. It is a key metric for assessing the overall goodness-of-fit of the model. A higher R² value means that the model explains a large portion of the variability in stock prices, making it a more reliable tool for forecasting. For business impact, a high R² ensures that the model captures the underlying trend and patterns in the stock prices, aiding in better investment strategies and risk management.

Both MSE and R² are critical for a positive business impact because they collectively ensure the accuracy and reliability of the model's predictions, which are fundamental for making informed financial decisions in the volatile stock market.
